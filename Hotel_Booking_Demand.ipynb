```python
import kagglehub
path = kagglehub.dataset_download("jessemostipak/hotel-booking-demand")
```

```python
import pandas as pd

df = pd.read_csv(path + "/hotel_bookings.csv")
df.head()
```

# Understanding Data

```python
df.info()
df.describe()
```

# Data Cleaning


```python
# Missing Value
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna('Unknown')
df['agent'] = df['agent'].fillna(0)
df = df.drop('company', axis=1)

print("Missing values after handling:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Duplicate Data
print(f"Number of duplicate rows before removal: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"Number of duplicate rows after removal: {df.duplicated().sum()}")
```

```python
# Correct Data Types
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df['agent'] = df['agent'].astype(int)
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month'].astype(str))
df = df.drop(columns=['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'])

print("Data types after correction:")
print(df[['reservation_status_date', 'agent', 'arrival_date']].dtypes)
```

```python
# Check the Outlier
print(df['adr'].describe())
```

```python
# Handle Outlier
df = df[df['adr'] > 0]
df = df[df['adr'] < 5000]

print(f"Shape of DataFrame after outlier removal: {df.shape}")
print(df['adr'].describe())
```

```python
df.info()
```

# Revenue Feature Engineering

```python
# Total Revenue
df['total_stay_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_revenue'] = 0
df.loc[df['is_canceled'] == 0, 'total_revenue'] = df['total_stay_nights'] * df['adr']

print(df[['total_stay_nights', 'adr', 'is_canceled', 'total_revenue']].head())
```

```python
df['total_stay_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_revenue'] = 0.0  # Initialize as float to avoid FutureWarning
df.loc[df['is_canceled'] == 0, 'total_revenue'] = df['total_stay_nights'] * df['adr']

print(df[['total_stay_nights', 'adr', 'is_canceled', 'total_revenue']].head())
```

```python
# Total Guests
df['children'] = df['children'].astype(int)
df['total_guests'] = df['adults'] + df['children'] + df['babies']

print(df[['adults', 'children', 'babies', 'total_guests']].head())
```

```python
import numpy as np

# Revenue per Guest
df['rev_per_guest'] = np.where(df['total_guests'] == 0, 0, df['total_revenue'] / df['total_guests'])

print(df[['total_revenue', 'total_guests', 'rev_per_guest']].head())
```

```python
# Categorize Lead Time
bins = [-1, 7, 30, 90, df['lead_time'].max()]
labels = ['Last Minute', 'Short Term', 'Medium Term', 'Early Bird']
df['lead_time_category'] = pd.cut(df['lead_time'], bins=bins, labels=labels, right=True)

print(df[['lead_time', 'lead_time_category']].head())
```

```python
# Potential Revenue Loss
df['potential_revenue_loss'] = 0.0
df.loc[df['is_canceled'] == 1, 'potential_revenue_loss'] = df['total_stay_nights'] * df['adr']

print(df[['is_canceled', 'total_stay_nights', 'adr', 'potential_revenue_loss']].head())
```

```python
print(df[['total_revenue', 'total_guests', 'rev_per_guest', 'lead_time_category', 'potential_revenue_loss']].head())
```

### Feature Engineering Summary:

*   **Total Stay Nights (`total_stay_nights`)**: Created by summing `stays_in_weekend_nights` and `stays_in_week_nights` to represent the total duration of a booking.
*   **Total Revenue (`total_revenue`)**: Calculated as `total_stay_nights * adr` for non-canceled bookings (`is_canceled == 0`). Initialized as a float to prevent `FutureWarning` due to type incompatibility.
*   **Total Guests (`total_guests`)**: Created by summing `adults`, `children` (after casting to `int`), and `babies` to get the total number of people in a booking.
*   **Revenue Per Guest (`rev_per_guest`)**: Calculated as `total_revenue / total_guests`. Handled division by zero by setting `rev_per_guest` to 0 when `total_guests` is 0.
*   **Lead Time Category (`lead_time_category`)**: Categorized `lead_time` into 'Last Minute' (0-7 days), 'Short Term' (8-30 days), 'Medium Term' (31-90 days), and 'Early Bird' (>90 days) using `pd.cut`.
*   **Potential Revenue Loss (`potential_revenue_loss`)**: Calculated as `total_stay_nights * adr` for canceled reservations (`is_canceled == 1`), representing revenue that could have been earned. Set to 0 for non-canceled reservations.
